```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import root_mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
df = pd.read_csv("dailyActivity_merged.csv")
```

```python
df['ActivityDate'] = pd.to_datetime(df['ActivityDate'])
```

Total Distance / Total Steps = StepIntensity
Very Distance / Very Minutes = VeryActiveIntensity
Moderate Distance / Fairly Minutes = ModerateActiveIntensity
Light Distance / Light Minutes = LightActiveIntensity

```python
df['StepsIntensity'] = df['TotalSteps']/df['TotalDistance']
df['VeryActiveIntensity'] = df['VeryActiveDistance']/df['VeryActiveMinutes']
df['ModeratelyActiveIntensity'] = df['ModeratelyActiveDistance']/df['FairlyActiveMinutes']
df['LightActiveIntensity'] = df['LightActiveDistance']/df['LightlyActiveMinutes']
```

```python
df.corr()
```

```python
df.sample(10)
```

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.fillna(df.mean(numeric_only=True), inplace=True)
```

#### Working Database
<hr>
• df1 = All features except ActivitiyDate and Id <br>
• df2 = All pre-exisiting features except ActivityDate and Id <br>
• df3 = All newly-made features <br>
• df4 = All features with corr >= 0.3 <br>

\_ Note : Out of ['TotalSteps','TotalDistance','TrackerDistance'] only 'TotalDistance' would be considered as they have very high corr \_


#### Models to be Used
<hr>
• RandomForestRegressor => Handles Non-linear data verywell and less prone to overfitting <br>
• XGBRegressor => Often offers higher accuracy and supports regularization <br>
• Linear Regression <br>

```python
df1 = df.drop(columns=['Id','ActivityDate','TrackerDistance','TotalSteps'])
df2 = df.drop(columns=['Id','ActivityDate','StepsIntensity','VeryActiveIntensity','ModeratelyActiveIntensity','LightActiveIntensity','TrackerDistance','TotalSteps'])
df3 = df[['Calories','StepsIntensity','VeryActiveIntensity','ModeratelyActiveIntensity','LightActiveIntensity']]

def filter_features_by_correlation(df, target='Calories', threshold=0.3):
    """
    Returns a new DataFrame with features whose correlation with the target is >= threshold.

    Parameters:
    df (pd.DataFrame): Original DataFrame
    target (str): Target column name to correlate with
    threshold (float): Correlation threshold (absolute value)

    Returns:
    pd.DataFrame: New DataFrame with selected features and target column
    """
    # Ensure the target exists
    if target not in df.columns:
        raise ValueError(f"Target column '{target}' not found in DataFrame.")

    # Compute correlation with the target
    corr = df.corr(numeric_only=True)[target].abs()

    # Select features with correlation >= threshold
    selected_features = corr[corr >= threshold].index.tolist()

    # Return new DataFrame with selected features
    return df[selected_features]

df4 = filter_features_by_correlation(df, target='Calories', threshold=0.3)
df4 = df4.drop(columns=['TotalSteps','TrackerDistance']) 
```

```python
X1 = df1.drop(columns=['Calories'])
y1 = df1['Calories']

X2 = df2.drop(columns=['Calories'])
y2 = df2['Calories']

X3 = df3.drop(columns=['Calories'])
y3 = df3['Calories']

X4 = df4.drop(columns=['Calories'])
y4 = df4['Calories']
```

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2, random_state=42)
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2, random_state=42)
X3_train, X3_test, y3_train, y3_test = train_test_split(X3, y3, test_size=0.2, random_state=42)
X4_train, X4_test, y4_train, y4_test = train_test_split(X4, y4, test_size=0.2, random_state=42)
```

##### Linear Regression on data

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
#Linear Regression Model for df1
lr.fit(X1_train, y1_train)
y1_pred = lr.predict(X1_test)
y1_rmse = root_mean_squared_error(y1_test, y1_pred)
y1_mae = mean_absolute_error(y1_test, y1_pred)
y1_r2 = r2_score(y1_test, y1_pred)
print(f"RMSE: {y1_rmse}, MAE: {y1_mae}, R2: {y1_r2}")
```

```python
#Linear Regression Model for df4
lr.fit(X4_train, y4_train)
y4_pred = lr.predict(X4_test)
y4_rmse = root_mean_squared_error(y4_test, y4_pred)
y4_mae = mean_absolute_error(y4_test, y4_pred)
y4_r2 = r2_score(y4_test, y4_pred)
print(f"RMSE: {y4_rmse}, MAE: {y4_mae}, R2: {y4_r2}")
```

##### RandomForestRegressor on data

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=8, random_state=42)
```

```python
#Random Forest Model for df1
rf.fit(X1_train, y1_train)
y1_pred_rf = rf.predict(X1_test)
y1_rmse_rf = root_mean_squared_error(y1_test, y1_pred_rf)
y1_mae_rf = mean_absolute_error(y1_test, y1_pred_rf)
y1_r2_rf = r2_score(y1_test, y1_pred_rf)
print(f"Random Forest RMSE: {y1_rmse_rf}, MAE: {y1_mae_rf}, R2: {y1_r2_rf}")
```

```python
#Random Forest Model for df2
rf.fit(X2_train, y2_train)
y2_pred_rf = rf.predict(X2_test)
y2_rmse_rf = root_mean_squared_error(y2_test, y2_pred_rf)
y2_mae_rf = mean_absolute_error(y2_test, y2_pred_rf)
y2_r2_rf = r2_score(y2_test, y2_pred_rf)
print(f"Random Forest RMSE: {y2_rmse_rf}, MAE: {y2_mae_rf}, R2: {y2_r2_rf}")
```

```python
rf.fit(X4_train, y4_train)
y4_pred_rf = rf.predict(X4_test)
y4_rmse_rf = root_mean_squared_error(y4_test, y4_pred_rf)
y4_mae_rf = mean_absolute_error(y4_test, y4_pred_rf)
y4_r2_rf = r2_score(y4_test, y4_pred_rf)
print(f"Random Forest RMSE: {y4_rmse_rf}, MAE: {y4_mae_rf}, R2: {y4_r2_rf}")
```

##### Using grid search cv

```python
param_grid_rf = {
    'n_estimators': [5,10,50,100,200,300,500],
    'max_depth': [None,5,10,15,20,25],
    'min_samples_split': [2,3, 5],
    'min_samples_leaf': [1, 2],
    'bootstrap': [True, False]
}

# Set up GridSearchCV
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid_rf,
    cv=5,                # 5-fold cross-validation
    scoring='neg_mean_squared_error',  # You can also use 'r2', 'neg_mean_absolute_error', etc.
    n_jobs=-1,
    verbose=1
)
```

```python
grid_search.fit(X1_train, y1_train)
best_rf = grid_search.best_estimator_
best_score_rf = grid_search.best_score_
print(f"Best parameters: {best_rf}")
print(f"Best score: {best_score_rf}")
```
